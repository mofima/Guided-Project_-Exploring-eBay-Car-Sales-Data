Further investigations reveals that the data in the price column as well in the odometer column are numeric but stored as strings.

```python
autos['price'] = autos['price'].str.replace("$", "").str.replace(",","")
autos['price'] = autos['price'].astype(int)
autos['price'].unique()
```

```python
autos['odometer'] = autos['odometer'].str.replace(",", "").str.replace("km","")
autos['odometer'] = autos['odometer'].astype(int)
autos['odometer'].unique()
```

```python
autos.rename({"odometer":"odometer_km"}, axis=1, inplace=True)
```

```python
autos = autos[autos["price"].between(1,720000)]
```

```python
# the following columns have no importance in the prediction, hence dropped
autos.drop(['date_crawled','ad_created','last_seen', 'name'], axis=1, inplace=True)
```

```python
data_cols = ['abtest', 'vehicle_type', 'registration_year', 'gearbox', 'power_ps', 'model', 'odometer_km','fuel_type', 'brand','unrepaired_damage']
data = autos[data_cols]

target = autos['price']
```

```python
data['age'] = 2022 - data['registration_year']
```

```python
data.drop(['registration_year'],axis=1,inplace=True)
```

```python
data['unrepaired_damage'].nunique()
```

```python
data.head()
```

```python
from sklearn.compose import make_column_selector as selector

cat_cols_selector = selector(dtype_include=object)
cat_cols = cat_cols_selector(data)

numcols_selector = selector(dtype_exclude=object)
num_cols = numcols_selector(data)
```

```python
num_cols, cat_cols
```

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OrdinalEncoder
from sklearn.linear_model import RidgeCV
import numpy as np
from sklearn.model_selection import cross_validate

cat_transformer = Pipeline([
                   ('imputer', SimpleImputer(strategy='most_frequent')),
                    ('Ordinal', OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1))
])

num_transformer = Pipeline([
                   ('imputer', SimpleImputer(strategy='most_frequent')),
                    ('scaler', StandardScaler())
])

preprocessor = ColumnTransformer([
                ('cat', cat_transformer, cat_cols),
                ('num', num_transformer, num_cols) 
])

alphas = np.logspace(-2,2,num=20)
model = Pipeline([
          ('preprocessor', preprocessor),
          ('regressor', RidgeCV(alphas=alphas))
])

```

```python
cv_results = cross_validate(model, data, target, cv=10, scoring='neg_mean_absolute_error')
cv_results = -cv_results['test_score']
cv_results.mean()
```

```python
from sklearn.ensemble import RandomForestRegressor
model = Pipeline([
          ('preprocessor', preprocessor),
          ('random_forest', RandomForestRegressor(random_state=0))
])

cv_results2 = cross_validate(model, data, target, cv=10, scoring='neg_mean_absolute_error')
cv_results2 = -cv_results2['test_score']
cv_results2.mean()
```
